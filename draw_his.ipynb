```python
import matplotlib.pyplot as plt
from src.utils import *
```

```python

# Đường dẫn đến ảnh
smoking_images = ["data/demo_img/smoking_0071.jpg", "data/demo_img/smoking_0246.jpg"]
notsmoking_images = ["data/demo_img/notsmoking_0406.jpg", "data/demo_img/notsmoking_0393.jpg"]
wrong_smoking = ["data/demo_img/smoking2.jpg","data/demo_img/smoking_0176.jpg"]

# Tham số histogram
# bin_h, bin_s, bin_v = 16, 16, 16  # Số lượng bin cho mỗi kênh HSV
bin = 128

# Hàm vẽ histogram
def plot_histogram(image_paths, title):
    plt.figure(figsize=(18, 5))

    for i, img_path in enumerate(image_paths):
        # Đọc ảnh và tính histogram
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  # Chuyển ảnh sang kênh HSV
        img = cv2.resize(img, (512, 512))
        hist = extract_color_histogram_hsv(img, bin)

        # Chia histogram thành 3 kênh: H, S, V
        # hist_h = hist[:bin]
        # hist_s = hist[bin:bin + bin]
        # hist_v = hist[bin + bin:]

        # Vẽ biểu đồ
        plt.subplot(1, len(image_paths), i + 1)
        # plt.plot(hist_h, color='r', label='Hue')
        # plt.plot(hist_s, color='g', label='Saturation')
        # plt.plot(hist_v, color='b', label='Value')
        plt.plot(hist, color='b', label='Feature vector')
        plt.title(f"{title} - Image {i+1}")
        plt.legend()
        plt.xlabel("Bins")
        plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()
```

```python
# Vẽ histogram cho Smoking
plot_histogram(smoking_images, "Smoking")
```

```python
# Vẽ histogram cho NotSmoking
plot_histogram(wrong_smoking, "Wrong predictions for smoking")
```

```python
# Vẽ histogram cho NotSmoking
plot_histogram(notsmoking_images, "NotSmoking")
```
